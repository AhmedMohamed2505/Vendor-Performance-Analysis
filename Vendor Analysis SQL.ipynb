- After reviewing separate summaries of purchasing and sales traffic, we will merge these tables into one comprehensive table.

```python
final_table = pd.read_sql_query("""WITH FreightSummary as (select VendorNumber, VendorName, sum(Freight) as FreightCost 
                                 from vendor_invoice 
                                 GROUP BY VendorNumber, VendorName
                                ),
                                PurchasesSummary as (select
                                  p.VendorName,
                                  p.VendorNumber,
                                  p.Description,
                                  p.Brand,
                                  p.PurchasePrice,
                                  pp.Volume,
                                  pp.Price as ActualPrice,
                                  SUM(p.Quantity) as TotalPurchasesQuantity,
                                  SUM(p.Dollars) as TotalPurchasesDollars
                                  FROM purchases p
                                  JOIN purchase_prices pp 
                                  ON p.Brand = pp.Brand 
                                  WHERE p.PurchasePrice > 0
                                  GROUP BY p.VendorName, p.VendorNumber, p.Brand
                                ),
                                SalesSummary as (select
                                        s.VendorName,
                                        s.VendorNo,
                                        s.Brand,
                                        s.SalesDate,
                                        SUM(SalesDollars) AS TotalSalesDollars,
                                        SUM(SalesPrice) AS TotalSalesPrice,
                                        SUM(SalesQuantity) AS TotalSalesQuantity,
                                        SUM(ExciseTax) AS TotalSalesTax
                                        FROM sales s
                                        GROUP BY VendorName, VendorNo, Brand
                                )
                               select 
                                ps.VendorName,
                                ps.VendorNumber,
                                ps.Description,
                                ps.Brand,
                                ps.PurchasePrice,
                                ps.ActualPrice,
                                ps.Volume,
                                ps.TotalPurchasesQuantity,
                                ps.TotalPurchasesDollars,
                                ss.TotalSalesDollars,
                                ss.SalesDate,
                                ss.TotalSalesPrice,
                                ss.TotalSalesQuantity,
                                ss.TotalSalesTax,
                                fs.FreightCost
                                FROM PurchasesSummary ps
                                LEFT JOIN SalesSummary ss
                                ON ps.VendorNumber = ss.VendorNo
                                AND ps.Brand = ss.Brand
                                LEFT JOIN FreightSummary fs
                                ON ps.VendorNumber = fs.VendorNumber
                                ORDER BY ps.TotalPurchasesDollars DESC""", conn)
final_table
```

- 1- Consolidation of Purchases, Sales, and Logistics

This step integrates Purchases, Sales, and Freight into a single Master Table. By combining these datasets, we can evaluate the full lifecycle 
of each brand and calculate the total profitability in one view.

- 2- Investment-Centric Analysis with LEFT JOIN

We use a LEFT JOIN on the Purchases table and sort by TotalPurchasesDollars (DESC) to focus on our highest capital investments. This approach ensures we track all inventoryeven unsold itemsto identify dead stock and prioritize our largest vendors.

```python
import pandas as pd
import os
from sqlalchemy import create_engine
import logging
import time
import sqlite3
from csv_to_db import ingest_db
 
logging.basicConfig(
    filename="logs/Script vendor_summary from new data.log",
    level=logging.DEBUG,
    format="%(asctime)s - %(levelname)s - %(message)s",
    filemode="a"
)

def create_vendor_summary(conn):

    final_table = pd.read_sql_query("""WITH FreightSummary as (select VendorNumber, VendorName, sum(Freight) as FreightCost 
                                 from vendor_invoice 
                                 GROUP BY VendorNumber, VendorName
                                ),
                                PurchasesSummary as (select
                                  p.VendorName,
                                  p.VendorNumber,
                                  p.Brand,
                                  p.PurchasePrice,
                                  pp.Volume,
                                  pp.Price as ActualPrice,
                                  SUM(p.Quantity) as TotalPurchasesQuantity,
                                  SUM(p.Dollars) as TotalPurchasesDollars
                                  FROM purchases p
                                  JOIN purchase_prices pp 
                                  ON p.Brand = pp.Brand 
                                  WHERE p.PurchasePrice > 0
                                  GROUP BY p.VendorName, p.VendorNumber, p.Brand
                                ),
                                SalesSummary as (select
                                        s.VendorName,
                                        s.VendorNo,
                                        s.Brand,
                                        SUM(SalesDollars) AS TotalSalesDollars,
                                        SUM(SalesPrice) AS TotalSalesPrice,
                                        SUM(SalesQuantity) AS TotalSalesQuantity,
                                        SUM(ExciseTax) AS TotalSalesTax
                                        FROM sales s
                                        GROUP BY VendorName, VendorNo, Brand
                                )
                               select 
                                ps.VendorName,
                                ps.VendorNumber,
                                ps.Brand,
                                ps.PurchasePrice,
                                ps.ActualPrice,
                                ps.Volume,
                                ps.TotalPurchasesQuantity,
                                ps.TotalPurchasesDollars,
                                ss.TotalSalesDollars,
                                ss.TotalSalesPrice,
                                ss.TotalSalesQuantity,
                                ss.TotalSalesTax,
                                fs.FreightCost
                                FROM PurchasesSummary ps
                                LEFT JOIN SalesSummary ss
                                ON ps.VendorNumber = ss.VendorNo
                                AND ps.Brand = ss.Brand
                                LEFT JOIN FreightSummary fs
                                ON ps.VendorNumber = fs.VendorNumber
                                ORDER BY ps.TotalPurchasesDollars DESC""", conn)
    

    return final_table

def cleaning_vendor_summary(final_table):

    final_table['VendorName'] = final_table['VendorName'].str.strip()
    final_table['Volume'] = final_table['Volume'].astype('float64')
    final_table.fillna(0, inplace=True)

    final_table['GrossProfit'] = final_table['TotalSalesDollars'] - final_table['TotalPurchasesDollars']
    final_table['ProfitMargin'] = final_table['GrossProfit'] / final_table['TotalSalesDollars']*100
    final_table['StockTurnOver'] = final_table['TotalSalesQuantity'] / final_table['TotalPurchasesQuantity'] 
    final_table['SalestoPurchaseRatio'] = final_table['TotalSalesDollars'] / final_table['TotalPurchasesDollars']

    return final_table

if __name__ == "__main__":

    conn = sqlite3.connect('inventory.db')

    logging.info('Creating Summary Table .... ')
    summary_df = create_vendor_summary(conn)
    logging.info(summary_df.head())

    logging.info('Cleaning Summary Table .... ')
    clean_df = cleaning_vendor_summary(summary_df)
    logging.info(clean_df.head())

    logging.info('Ingesting Data .... ')
    clean_df.to_sql('final_table', conn, if_exists='replace', index=False)
    logging.info('Completed')
```
